```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/kc_house_data_cleaned.csv', parse_dates=['date'])
```

```python
df.dtypes
```

```python
categorical_feats = ['bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'condition',
                     'grade', 'zipcode', 'yr_built', 'yr_renovated']
```

```python
continuous_feats = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built', 
                    'yr_renovated', 'sqft_living15', 'sqft_lot15']
```

```python
for i in categorical_feats:
    plt.figure(figsize=(15,10))
    sns.boxplot(x=df[i], y=df.price)
    plt.show()
```

```python
for i in continuous_feats:
    plt.figure(figsize=(15,10))
    sns.scatterplot(x=df[i], y=df.price)
    plt.show()
```
